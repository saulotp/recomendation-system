## colaborative filter in user based

```python
## import libraries
import pandas as pd
import numpy as np
```

```python
## import data
movie_df = pd.read_csv('movies.csv', sep=',')
movie_df
```

we wont use the genres column, then we can select only the first 2 columns

```python
movie_df = movie_df[["movieId", "title"]]
movie_df
```

```python
ratings_df = pd.read_csv('ratings.csv', sep=',')
ratings_df
```

```python
ratings_df = ratings_df[["userId", "movieId", "rating"]]
ratings_df
```

```python
main_df = pd.merge(movie_df, ratings_df, on='movieId')
main_df
```

```python
recomend_df = main_df.pivot_table(index='userId', columns='title', values='rating').fillna(0)
recomend_df
```
