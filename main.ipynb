## colaborative filter in user based

```python
## import libraries
import pandas as pd
import numpy as np
```

```python
## import data
movie_df = pd.read_csv('movies.csv', sep=',')
movie_df
```

we wont use the genres column, then we can select only the first 2 columns

```python
movie_df = movie_df[["movieId", "title"]]
movie_df
```

```python
ratings_df = pd.read_csv('ratings.csv', sep=',')
ratings_df
```

```python
ratings_df = ratings_df[["userId", "movieId", "rating"]]
ratings_df
```

```python
main_df = pd.merge(movie_df, ratings_df, on='movieId')
main_df
```

```python
recomend_df = main_df.pivot_table(index='userId', columns='title', values='rating').fillna(0)
recomend_df
```

```python
from sklearn.neighbors import NearestNeighbors

knn_model = NearestNeighbors(metric='cosine')
knn_model.fit(recomend_df)
```

```python
recomend_df.iloc[0]
recomend_df.index[0]
```

```python
user_index = 230
distance, neighbors_index = knn_model.kneighbors(recomend_df.iloc[user_index].values.reshape(1, -1), n_neighbors=3)
```

```python
distance
```

```python
neighbors_index
```

```python
for i in range(0, len(distance.flatten())):
    if i == 0:
        print(f'Recommendations for user: {str(recomend_df.index[user_index]-1)}')
    else:
        print(f'Neighbor: {str(recomend_df.index[neighbors_index.flatten()[i]-1])} with distance: {str(distance.flatten()[i])}')
```
