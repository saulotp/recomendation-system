```python
import pandas as pd
import numpy as np

pcclient = pd.read_csv('pcclient.txt', sep='|')

pcclient.columns
pcclient.rename(columns = {'CLIENTE                                                    ':'CLIENTE'}, inplace = True)
pcclient = pcclient[["CODCLI", "CLIENTE"]]
pcclient = pcclient.drop([0], axis=0)
pcclient = pcclient.reset_index(drop=True)
#pcclient["CODCLI"] = pd.to_numeric(pcclient["CODCLI"])
#pcclient = pcclient.sort_values(by=['CODCLI'], ascending=True)

pcclient.to_csv('pcclient.csv', index=False)
```

```python
pcprodut = pd.read_csv('pcprodut.txt', sep='|')
pcprodut.columns    
pcprodut.rename(columns = {'DESCRICAO                               ':'DESCRICAO'}, inplace = True)
pcprodut = pcprodut[["CODPROD", "DESCRICAO"]]
pcprodut = pcprodut.drop([0], axis=0)
pcprodut = pcprodut.reset_index(drop=True)
pcprodut
```

```python
pcpedi  = pd.read_csv('pcpedi.txt', sep='|')
pcpedi.columns
pcpedi.rename(columns = {'DATA                   ':'DATA', 'NUMPED       ':'NUMPED', 'QT   ': 'QT'}, inplace = True)
pcpedi = pcpedi[["DATA", "NUMPED", "QT", "CODPROD", "CODCLI"]]
pcpedi = pcpedi.drop([0], axis=0)
pcpedi = pcpedi.reset_index(drop=True)
pcpedi
```

```python
main = pcpedi.drop(['DATA', 'NUMPED'], axis=1)
main
```

```python
recomend_df = main.pivot_table(index=['CODPROD'], columns=['CODCLI'], values=['QT'], aggfunc=np.sum)
recomend_df = recomend_df.fillna(0)
recomend_df
```
