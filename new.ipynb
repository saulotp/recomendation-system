```python
import pandas as pd
import numpy as np

pcclient = pd.read_csv('pcclient.csv', sep=';')
pcprodut = pd.read_csv('pcprodut.csv', sep=';')
pcpedi  = pd.read_csv('pcpedi.csv', sep=';')




```

```python

list = []
count = 0
for i in pcclient["CODCLI"]:    
    list.append(count)
    count += 1

tst = pd.DataFrame(list)
tst.columns = ["NEWCODCLI"]

pcclient = pd.concat([pcclient, tst], axis=1)
pcclient
```

```python
main = pd.merge(pcpedi, pcprodut, on='CODPROD')
main
```

```python
from scipy.sparse import csr_matrix
recomend_df = main.pivot_table(index=['NEWCODPROD'], columns=['CODCLI'], values=['QT'], aggfunc=np.sum)
recomend_df = recomend_df.fillna(0)
#recomend_df = csr_matrix( recomend_df )
recomend_df
```

```python
from sklearn.neighbors import NearestNeighbors


knn_model = NearestNeighbors(metric='cosine')
knn_model.fit(recomend_df)
```

```python
recomend_df
#print(recomend_df[recomend_df["QT"]==1].index.values)
#print(recomend_df.index[recomend_df['CODCLI'] == 3100].tolist())
```

```python
recomend_df.index
```

```python
idx_tst = 3
distance, neighbors_index = knn_model.kneighbors(recomend_df.loc[idx_tst].values.reshape(1, -1), n_neighbors=6)
```

```python
idx_tst
```

```python
neighbors_index
```

```python


for i in range(0, len(distance.flatten())):
    if i == 0:
        print(f'Recommendations for user: {str(recomend_df.index[idx_tst]-1)}')
    else:
        print(f'Neighbor: {str(recomend_df.index[neighbors_index.flatten()[i]])} with distance: {str(distance.flatten()[i])}')
```
