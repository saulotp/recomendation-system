```python
import pandas as pd
import numpy as np

pcclient = pd.read_csv('pcclient.csv', sep=';')
pcprodut = pd.read_csv('pcprodut.csv', sep=';')
pcpedi  = pd.read_csv('pcpedi.csv', sep=';')
```

```python
list = []
count = 0
for i in pcclient["CODCLI"]:    
    list.append(count)
    count += 1

tst = pd.DataFrame(list)
tst.columns = ["NEWCODCLI"]

pcclient = pd.concat([pcclient, tst], axis=1)
pcclient
```

```python
main = pd.merge(pcpedi, pcprodut, on='CODPROD')
"""merged_df2 = pd.merge(merged_df, pcclient, on='CODCLI')
full_df = merged_df2.drop(['CLIENTE', 'CODCLI', 'DESCRICAO'], axis=1)
main = full_df.sort_values(by=['NEWCODCLI'], ascending=True)"""
main
```

```python
matrix_df = main.pivot_table(index=['CODCLI'], columns=['DESCRICAO'], values=['QT'], aggfunc=np.sum).fillna(0)
recomend_df = matrix_df
recomend_df = recomend_df.reset_index(drop=True)
```

```python
recomend_df
```

```python
from sklearn.neighbors import NearestNeighbors


knn_model = NearestNeighbors(metric='cosine')
knn_model.fit(recomend_df)
```

```python
recomend_df.index[0]
```

```python
idx_user = 3100
```

```python
idx_user = 2
distance, neighbors_index = knn_model.kneighbors(recomend_df.iloc[idx_user].values.reshape(1, -1), n_neighbors=5)
```

```python
distance.flatten()
```

```python
neighbors_index.flatten()
```

```python
user = recomend_df.index[idx_user]

for i in range(0, len(distance.flatten())):
    if i == 0:
        print(f'Recommendations for user: {user}')
    else:
        print(f'Neighbor: {str(recomend_df.index[neighbors_index.flatten()[i]])} with distance: {str(distance.flatten()[i])}')
```

```python
ds_user = recomend_df.loc[user].to_frame()
neighbors_index_nearest = recomend_df.index[neighbors_index.flatten()[1]]
#ds_user
ds_neighbor = recomend_df.loc[neighbors_index_nearest].to_frame()
```

```python
result = pd.merge(ds_user, ds_neighbor, on='DESCRICAO').sort_values(by=neighbors_index_nearest)
result
```

```python
result = result[(result[user] > 0) & (result[neighbors_index_nearest] == 0)]
result
```
